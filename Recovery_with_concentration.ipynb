```python
df_combined = pd.concat([Sonication_1_1, Sonication_1_2, Sonication_1_3, Sonication_2_1, Sonication_2_2, Sonication_2_3, Sonication_3_1, Sonication_3_2, Sonication_3_3, Sonication_Ref_1, Sonication_Ref_2, Sonication_Ref_3])

analyte_mapping = {analyte: i for i, analyte in enumerate(df_combined['Analyte'].unique())}
df_combined['AnalyteNumeric'] = df_combined['Analyte'].map(analyte_mapping)

analyte_mean = df_combined.groupby(['Series', 'AnalyteNumeric'])['Concentration [microg/ml]'].mean().reset_index()

series_dataframes = {}
for series in df_combined['Series'].unique():
    series_df = analyte_mean[analyte_mean['Series'] == series]
    series_dataframes[series] = series_df

for series, series_df in series_dataframes.items():
    print(f"Mean for Series {series}:")
    print(series_df)
    print("\n")
```

```python
analyte_std = df_combined.groupby(['Series', 'AnalyteNumeric'])['Concentration [microg/ml]'].std().reset_index()

series_dataframes_std = {}
for series in df_combined['Series'].unique():
    series_df = analyte_mean[analyte_mean['Series'] == series]
    series_dataframes_std[series] = series_df

for series, series_df in series_dataframes_std.items():
    print(f"Std for Series {series}:")
    print(series_df)
    print("\n")
```

```python
analyte_mapping = {analyte: i for i, analyte in enumerate(df_combined['Analyte'].unique())}
df_combined['AnalyteNumeric'] = df_combined['Analyte'].map(analyte_mapping)

analyte_mean = df_combined.groupby(['Series', 'AnalyteNumeric'])['Concentration [microg/ml]'].mean().reset_index()
analyte_std = df_combined.groupby(['Series', 'AnalyteNumeric'])['Concentration [microg/ml]'].std().reset_index()

series_dataframes_mean = {}
series_dataframes_std = {}
for series in df_combined['Series'].unique():
    series_mean_df = analyte_mean[analyte_mean['Series'] == series]
    series_std_df = analyte_std[analyte_std['Series'] == series]
    series_dataframes_mean[series] = series_mean_df
    series_dataframes_std[series] = series_std_df

plt.figure(figsize=(10, 6))
for series, series_mean_df in series_dataframes_mean.items():
    series_std_df = series_dataframes_std[series]
    plt.errorbar(series_mean_df['AnalyteNumeric'], series_mean_df['Concentration [microg/ml]'],
                 yerr=series_std_df['Concentration [microg/ml]'], fmt='o', label=f'Series {series}', capsize=5)

plt.xlabel('Analyte')
plt.ylabel('Concentration (Mean)')
plt.title('Concentration Mean with Standard Deviation for Different Series')
plt.legend(loc='upper right')
plt.grid(True, axis='y', alpha=0.7)

plt.show()
```

```python
analyte_mean = df_combined.groupby(['Series', 'AnalyteNumeric'])['Concentration [microg/ml]'].mean().reset_index()

pivoted_data = analyte_mean.pivot(index='AnalyteNumeric', columns='Series', values='Concentration [microg/ml]')

pivoted_data['Ratio_1_to_Ref'] = pivoted_data[1] / pivoted_data['Ref'] * 100
pivoted_data['Ratio_2_to_Ref'] = pivoted_data[2] / pivoted_data['Ref'] * 100
pivoted_data['Ratio_3_to_Ref'] = pivoted_data[3] / pivoted_data['Ref'] * 100

pivoted_data.reset_index(inplace=True)

ratios_df = pivoted_data[['AnalyteNumeric', 'Ratio_1_to_Ref', 'Ratio_2_to_Ref', 'Ratio_3_to_Ref']]

#Optionally, replace or drop any infinite or NaN values in the ratios_df
ratios_df.replace([np.inf, -np.inf], np.nan, inplace=True)
ratios_df.dropna(inplace=True)

print("Ratios:")
print(ratios_df)
```

```python
analyte_std = df_combined.groupby(['Series', 'AnalyteNumeric'])['Concentration [microg/ml]'].std().reset_index()

pivoted_data_std = analyte_std.pivot(index='AnalyteNumeric', columns='Series', values='Concentration [microg/ml]')

pivoted_data_std['Ratio_1_to_Ref'] = (pivoted_data[1] / pivoted_data['Ref']) * 100 * np.sqrt(
    (pivoted_data_std[1] / pivoted_data[1])**2 + (pivoted_data_std['Ref'] / pivoted_data['Ref'])**2
)

pivoted_data_std['Ratio_2_to_Ref'] = (pivoted_data[2] / pivoted_data['Ref']) * 100 * np.sqrt(
    (pivoted_data_std[2] / pivoted_data[2])**2 + (pivoted_data_std['Ref'] / pivoted_data['Ref'])**2
)

pivoted_data_std['Ratio_3_to_Ref'] = (pivoted_data[3] / pivoted_data['Ref']) * 100 * np.sqrt(
    (pivoted_data_std[3] / pivoted_data[3])**2 + (pivoted_data_std['Ref'] / pivoted_data['Ref'])**2
)

pivoted_data_std.reset_index(inplace=True)

ratios_std_df = pivoted_data_std[['AnalyteNumeric', 'Ratio_1_to_Ref', 'Ratio_2_to_Ref', 'Ratio_3_to_Ref']]

ratios_std_df.replace([np.inf, -np.inf], np.nan, inplace=True)
ratios_std_df.dropna(inplace=True)

print("Ratios:")
print(ratios_std_df)
```

```python
fig, ax = plt.subplots()

for col in ratios_df.columns[1:]:
    mean_values = ratios_df[col]
    std_values = ratios_std_df[col]
    print(mean_values)
    print(std_values)

    ax.scatter(ratios_df['AnalyteNumeric'], mean_values, label=col)

    ax.errorbar(ratios_df['AnalyteNumeric'], mean_values, yerr=std_values, fmt='o', capsize=5, label=None)

ax.set_xlabel('Analyte Numeric')
ax.set_ylabel('Ratio (Percentage)')
ax.set_title('Scatter Plot with Error Bars')

ax.legend()

plt.figure(figsize=(10, 6))
plt.show()
```
