```python
dfs = [('Auto_5_gp_df', Auto_5_gp_df), ('Auto_4_gp_df', Auto_4_gp_df),
       ('Auto_3_gp_df', Auto_3_gp_df), ('Auto_2_gp_df', Auto_2_gp_df),
       ('Auto_1_gp_df', Auto_1_gp_df), ('Zafira_2_gp_df', Zafira_2_gp_df),
       ('Zafira_1_gp_df', Zafira_1_gp_df), ('Auto_5_pm_df', Auto_5_pm_df),
       ('Auto_4_pm_df', Auto_4_pm_df), ('Auto_3_pm_df', Auto_3_pm_df),
       ('Auto_2_pm_df', Auto_2_pm_df), ('Auto_1_pm_df', Auto_1_pm_df),
       ('Zafira_2_pm_df', Zafira_2_pm_df), ('Zafira_1_pm_df', Zafira_1_pm_df),
       ('U181_60k_df', U181_60k_df), ('U181_120k_df', U181_120k_df),
       ('U166_df', U166_df), ('U164_df', U164_df)]
```

```python
# Create a list of all unique compounds
all_compounds = []
for _, df in dfs:
    all_compounds.extend(df['Component Name'].unique())

# Remove duplicates and create an empty dataframe with all compounds and samples
all_compounds = list(set(all_compounds))
compound_sample_df = pd.DataFrame(index=all_compounds)

# Fill the dataframe with area ratios
for df_name, df in dfs:
    for compound in all_compounds:
        # Check if the compound is in the current dataframe
        if compound in df['Component Name'].values:
            # Get the area ratio for the compound in the current sample
            area_ratio = df.loc[df['Component Name'] == compound, 'Area Ratio'].values[0]
            compound_sample_df.loc[compound, df_name] = area_ratio
        else:
            # If compound is not in the current sample, set the value to 0
            compound_sample_df.loc[compound, df_name] = 0

# Display the resulting dataframe
print(compound_sample_df)
```

```python
all_compounds = []
for _, df in dfs:
    all_compounds.extend(df['Component Name'].unique())

# Remove duplicates and create an empty dataframe with all compounds and samples
all_compounds = list(set(all_compounds))
compound_sample_df = pd.DataFrame(index=all_compounds)

# Add a new column for molecule formula
compound_sample_df['Formula (mol ion)'] = ''

# Fill the dataframe with area ratios and molecule formula
for df_name, df in dfs:
    for compound in all_compounds:
        # Check if the compound is in the current dataframe
        if compound in df['Component Name'].values:
            # Get the area ratio for the compound in the current sample
            area_ratio = df.loc[df['Component Name'] == compound, 'Area Ratio'].values[0]
            # Get the molecule formula for the compound in the current sample
            formula = df.loc[df['Component Name'] == compound, 'Formula (mol ion)'].values[0]

            # Set the values in the compound_sample_df
            compound_sample_df.loc[compound, df_name] = area_ratio
            compound_sample_df.loc[compound, 'Formula (mol ion)'] = formula
        else:
            # If compound is not in the current sample, set the value to 0 for area ratio
            compound_sample_df.loc[compound, df_name] = 0

# Display the resulting dataframe
print(compound_sample_df)
```

```python
compound_sample_df_rearanged = compound_sample_df.sort_values(by='Formula (mol ion)')

print(compound_sample_df_rearanged)
```

```python
duplicated_rows = compound_sample_df[compound_sample_df_rearanged.duplicated(subset='Formula (mol ion)', keep=False)]

# Display the resulting dataframe with duplicated rows
print(duplicated_rows)
```

```python
combined_duplicated_rows = pd.DataFrame(columns=compound_sample_df.columns)

# Iterate through the duplicated rows
for formula, group in duplicated_rows.groupby('Formula (mol ion)'):
    # Determine which row to keep based on the number of 0 values
    selected_row = group.loc[group.eq(0).sum(axis=1).idxmax()]
    # Replace 0 values in the selected row with non-zero values from the other duplicated rows
    for idx, row in group.iterrows():
        selected_row = selected_row.where(selected_row.ne(0) | row.ne(0), other=row)

    # Add the selected row to the combined dataframe
    combined_duplicated_rows = combined_duplicated_rows.append(selected_row)

# Display the resulting dataframe with combined duplicated rows
print(combined_duplicated_rows)
```
