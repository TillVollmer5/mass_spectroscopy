```python
dfs = [('Auto_5_gp_df', Auto_5_gp_df), ('Auto_4_gp_df', Auto_4_gp_df),
       ('Auto_3_gp_df', Auto_3_gp_df), ('Auto_2_gp_df', Auto_2_gp_df),
       ('Auto_1_gp_df', Auto_1_gp_df), ('Zafira_2_gp_df', Zafira_2_gp_df),
       ('Zafira_1_gp_df', Zafira_1_gp_df), ('Auto_5_pm_df', Auto_5_pm_df),
       ('Auto_4_pm_df', Auto_4_pm_df), ('Auto_3_pm_df', Auto_3_pm_df),
       ('Auto_2_pm_df', Auto_2_pm_df), ('Auto_1_pm_df', Auto_1_pm_df),
       ('Zafira_2_pm_df', Zafira_2_pm_df), ('Zafira_1_pm_df', Zafira_1_pm_df),
       ('U181_60k_df', U181_60k_df), ('U181_120k_df', U181_120k_df),
       ('U166_df', U166_df), ('U164_df', U164_df)]
```

```python
all_compounds = []
for _, df in dfs:
    all_compounds.extend(df['Component Name'].unique())

# Remove duplicates and create an empty dataframe with all compounds and samples
all_compounds = list(set(all_compounds))
compound_sample_df = pd.DataFrame(index=all_compounds)

# Add a new column for molecule formula
compound_sample_df['Formula (mol ion)'] = ''

# Fill the dataframe with area ratios and molecule formula
for df_name, df in dfs:
    for compound in all_compounds:
        # Check if the compound is in the current dataframe
        if compound in df['Component Name'].values:
            # Get the area ratio for the compound in the current sample
            area_ratio = df.loc[df['Component Name'] == compound, 'Area Ratio'].values[0]
            # Get the molecule formula for the compound in the current sample
            formula = df.loc[df['Component Name'] == compound, 'Formula (mol ion)'].values[0]

            # Set the values in the compound_sample_df
            compound_sample_df.loc[compound, df_name] = area_ratio
            compound_sample_df.loc[compound, 'Formula (mol ion)'] = formula
        else:
            # If compound is not in the current sample, set the value to 0 for area ratio
            compound_sample_df.loc[compound, df_name] = 0

# Display the resulting dataframe
print(compound_sample_df)
```

```python
compound_sample_df_rearanged = compound_sample_df.sort_values(by='Formula (mol ion)')

print(compound_sample_df_rearanged)
```

```python
duplicated_rows = compound_sample_df[compound_sample_df.duplicated(subset='Formula (mol ion)', keep=False)]

# Display the resulting dataframe with duplicated rows
print(duplicated_rows)
```

```python
#Not the correct one, does not differentiate if two values in duplicates in the same column

combined_duplicated_rows = pd.DataFrame(columns=compound_sample_df.columns)

# Iterate through the duplicated rows
for formula, group in duplicated_rows.groupby('Formula (mol ion)'):
    # Determine which row to keep based on the number of 0 values
    selected_row = group.loc[group.eq(0).sum(axis=1).idxmax()]
    # Replace 0 values in the selected row with non-zero values from the other duplicated rows
    for idx, row in group.iterrows():
        selected_row = selected_row.where(selected_row.ne(0) | row.ne(0), other=row)

    # Add the selected row to the combined dataframe
    combined_duplicated_rows = combined_duplicated_rows.append(selected_row)

# Display the resulting dataframe with combined duplicated rows
print(combined_duplicated_rows)
combined_duplicated_rows.to_csv('/content/drive/My Drive/unknown_screening_data/Hitlist_non_dublicates_total_reduction.csv')
```

```python
combined_duplicated_rows = pd.DataFrame(columns=duplicated_rows.columns)

for formula, group in duplicated_rows.groupby('Formula (mol ion)'):
    # Initialize the selected row with the row having the minimum non-zero values
    selected_row = group.loc[group.ne(0).sum(axis=1).idxmin()].copy()

    # Iterate over columns to check the condition
    for col in group.columns[1:]:  # Skip the 'Formula (mol ion)' column
        # Check if there's a zero in the same column for the selected row
        if selected_row[col] == 0:
            # Debugging information
            print(f"Formula: {formula}, Column: {col}")

            # Find the row with the maximum non-zero values across all columns
            max_nonzero_row = group.loc[group.sum(axis=1).idxmax()]

            # Combine the rows if there are two non-zero values in the same column
            selected_row[col] = max_nonzero_row[col]

    # Add the selected row to the combined dataframe
    combined_duplicated_rows = combined_duplicated_rows.append(selected_row)

# Display the resulting dataframe with combined duplicated rows
print(combined_duplicated_rows)

# Save the resulting dataframe to a CSV file
combined_duplicated_rows.to_csv('/content/drive/My Drive/unknown_screening_data/Hitlist_non_duplicates_total_reduction.csv')
```

```python
compound_sample_df_no_duplicates = compound_sample_df.drop_duplicates(subset='Formula (mol ion)', keep=False)

# Concatenate the modified compound_sample_df with the combined_duplicated_rows DataFrame
final_result = pd.concat([compound_sample_df_no_duplicates, combined_duplicated_rows], ignore_index=False)

print(final_result.shape)

# Save the resulting dataframe to a CSV file
final_result.to_csv('/content/drive/My Drive/unknown_screening_data/Hitlist_non_duplicates_final.csv')
```

```python
detection_frequency = final_result.iloc[:, 1:].gt(0).sum(axis=1) / (final_result.shape[1] - 1)

# Insert the new column after the 'Formula (mol ion)' column
final_result.insert(1, 'Detection Frequency', detection_frequency)

print(final_result.head())

# Save the result to a new CSV file
final_result.to_csv('/content/drive/My Drive/unknown_screening_data/Result_With_Detection_Frequency.csv')
```

```python
detection_frequency_threshold = 0.2  # Adjust this threshold as needed

# Filter rows based on the detection frequency threshold
filtered_result = final_result[final_result['Detection Frequency'] >= detection_frequency_threshold]

print(filtered_result.shape)
# Save the filtered result to a new CSV file
filtered_result.to_csv('/content/drive/My Drive/unknown_screening_data/Filtered_Result_With_Detection_Frequency_02.csv')
```
