```python
#Boxplot of all the data from the gas phase results of the new cars, not correct values yet, works with lists... not pandas dataframes
sample_gp_new = [Auto_1_gp, Auto_2_gp, Auto_3_gp]

compound_data = {}

for element in sample_gp_new:
    for compound, value in element:
        if compound not in compound_data:
            compound_data[compound] = []

        compound_data[compound].append(value)

for compound, values in compound_data.items():
    print(f"{compound}: {values}")

compound_names = list(compound_data.keys())
compound_values = list(compound_data.values())

fig, ax = plt.subplots(figsize=(10, 6))

ax.boxplot(compound_values, labels=compound_names, showmeans=True, showfliers=True)
ax.set_title('Box-and-Whisker Plot for Compounds')
ax.set_ylabel('Second Data Point')
plt.tight_layout()
plt.xticks(rotation='vertical')
plt.show()
```
