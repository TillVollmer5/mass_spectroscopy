```python
df_combined = pd.concat([Auto_1_gp_df,Auto_2_gp_df, Auto_2_pm_df, Auto_3_gp_df, Auto_3_pm_df, Auto_4_gp_df, Auto_4_pm_df])

# Map unique 'Phthalat:' values to numerical values
phthalate_mapping = {phthalate: i for i, phthalate in enumerate(df_combined['Phthalat:'].unique())}

# Add a new column 'PhthalateNumeric' to the DataFrame based on the mapping
df_combined['PhthalateNumeric'] = df_combined['Phthalat:'].map(phthalate_mapping)

means = df_combined.groupby(['PhthalateNumeric', 'Option'])['Concentration [ng/m^3]'].mean().reset_index()


plt.figure(figsize=(16, 10))

ax = sns.boxplot(x='PhthalateNumeric', y='Concentration [ng/m^3]', hue='Option', data=df_combined, palette={'gp': 'skyblue', 'pm': 'lightcoral'}, showfliers=True)

sns.swarmplot(x='PhthalateNumeric', y='Concentration [ng/m^3]', hue='Option', data=df_combined, palette={'gp': 'skyblue', 'pm': 'lightcoral'}, size=5, dodge=True, edgecolor='black', linewidth=0.5)

for index, row in means.iterrows():
    x_offset = -0.2 if row['Option'] == 'gp' else 0.2
    ax.scatter(x=row['PhthalateNumeric'] + x_offset, y=row['Concentration [ng/m^3]'], color={'gp': 'black', 'pm': 'black'}[row['Option']], marker='x', s=100, linewidths=2)

plt.title('Phthalate concentration of new cars (n=3)')
plt.xlabel('Phthalate')
plt.ylabel('Concentration [ng/m^3]')
plt.xticks(ticks=range(len(phthalate_mapping)), labels=df_combined['Phthalat:'].unique())
plt.legend(loc='upper right')
plt.grid(True, axis='y', alpha=0.7)
plt.ylim(bottom=0)
plt.show()
```

```python
df_gps = [Auto_1_gp_df, Auto_2_gp_df, Auto_3_gp_df, Auto_4_gp_df]
df_pms = [Auto_1_pm_df, Auto_2_pm_df, Auto_3_pm_df, Auto_4_pm_df]

# Initialize mean_df_gp, std_df_gp, mean_df_pm, and std_df_pm using the 'Phthalat:' column from the first DataFrame
mean_df_gp = pd.DataFrame({'Phthalat:': df_gps[0]['Phthalat:']}).copy()
std_df_gp = pd.DataFrame({'Phthalat:': df_gps[0]['Phthalat:']}).copy()
mean_df_pm = pd.DataFrame({'Phthalat:': df_pms[0]['Phthalat:']}).copy()
std_df_pm = pd.DataFrame({'Phthalat:': df_pms[0]['Phthalat:']}).copy()

for i, df_gp in enumerate(df_gps):
    col_name = df_gp.columns[1] + f'_{i+1}'
    mean_df_gp[col_name] = df_gp.iloc[:, 1].copy()
    std_df_gp[col_name] = df_gp.iloc[:, 1].copy()
for i, df_pm in enumerate(df_pms):
    col_name = df_pm.columns[1] + f'_{i+1}'
    mean_df_pm[col_name] = df_pm.iloc[:, 1].copy()
    std_df_pm[col_name] = df_pm.iloc[:, 1].copy()

mean_df_gp['Mean'] = mean_df_gp.iloc[:, 1:].mean(axis=1)
std_df_gp['Std'] = std_df_gp.iloc[:, 1:].std(axis=1)
mean_df_pm['Mean'] = mean_df_pm.iloc[:, 1:].mean(axis=1)
std_df_pm['Std'] = std_df_pm.iloc[:, 1:].std(axis=1)

mean_df_gp = mean_df_gp[['Phthalat:', 'Mean']]
std_df_gp = std_df_gp[['Phthalat:', 'Std']]
mean_df_pm = mean_df_pm[['Phthalat:', 'Mean']]
std_df_pm = std_df_pm[['Phthalat:', 'Std']]

print(mean_df_gp)
print(std_df_gp)
print(mean_df_pm)
print(std_df_pm)

mean_df_gp['Option'] = 'gp'
mean_df_pm['Option'] = 'pm'

df_gp_combined = pd.concat([mean_df_gp, std_df_gp['Std']], axis=1)
df_pm_combined = pd.concat([mean_df_pm, std_df_pm['Std']], axis=1)

fig, ax = plt.subplots(figsize=(16, 10))
bar_width = 0.35
index_gp = np.arange(len(df_gp_combined))
index_pm = np.arange(len(df_pm_combined))

ax.bar(index_gp, df_gp_combined['Mean'], bar_width, yerr=df_gp_combined['Std'], label='gp', capsize=5)
ax.bar(index_pm + bar_width, df_pm_combined['Mean'], bar_width, yerr=df_pm_combined['Std'], label='pm', capsize=5)

ax.set_xlabel('Phthalat:')
ax.set_ylabel('Mean')
ax.set_xticks(index_gp + bar_width / 2)
ax.set_xticklabels(df_gp_combined['Phthalat:'])
plt.grid(True, axis='y', alpha=0.7)
ax.legend()
ax.set_ylim(bottom=0)
plt.show()
```
